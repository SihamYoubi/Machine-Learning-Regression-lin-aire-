# Advertissing problem avec sklearn

```python
 #Dataset
 import pandas as pnd
 import matplotlib.pyplot as plt
 from sklearn.model_selection import train_test_split
 from sklearn.linear_model import LinearRegression
 from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score 

 # Charger le dataset
 dataset=pnd.read_csv("Advertising.csv",index_col=0)
 # Sélection des trois premières colonnes du dataset : ce sont les variables d'entrée (features)
 X=dataset.iloc[:,:3]
 # Sélection de la quatrième colonne : c'est la variable cible à prédire (ici : Sales = Données )
 y=dataset.iloc[:,3]
 # Séparation du dataset en 2 parties (train et test)
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=55)

 # Création et entraînement du modèle
 model=LinearRegression() 
 model.fit(X_train,y_train) 
 # Prédiction sur les données de test
 yhat=model.predict(X_test) 
```

```python
# Paramètres du modèle
bias= model.intercept_
w = model.coef_
```

```python
print("Biais (intercept) :", bias)
print("Coefficients (w) :", w)
```

```python
mae = mean_absolute_error(y_test, yhat)
mse = mean_squared_error(y_test, yhat)
r2 = r2_score(y_test, yhat)
```

```python
print("Erreur Absolue Moyenne :", mae)
print("Erreur Quadratique Moyenne :", mse)
print("R2 Score :", r2)
```

```python
# Visualisation des résultats
plt.figure(figsize=(12, 6))
# Comparaison des valeurs réelles et prédites
plt.subplot(1, 2, 1)
plt.scatter(y_test, yhat, label='Prédictions',color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2, label='Ligne de Référence')
plt.xlabel('Données Réelles')
plt.ylabel('Données  Prédites')
plt.title('Données Réelles vs Prédites')
plt.legend()
```
