# Programme de prédiction de la consommation de boissons fraîches en fonction de la température.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd

# Jeu de données fictif : température et consommation de boissons
data = {
    'Température (°C)': [10, 15, 20, 25, 30],  # Températures en °C
    'Consommation (L)': [5, 8, 11, 15, 18]  # Consommation de boissons (L)
}

# Créer un DataFrame à partir des données
dataset = pd.DataFrame(data)

# Afficher le DataFrame
print(dataset)

# Extraire les variables d'entrée (température) et la variable cible (consommation)
X = dataset[['Température (°C)']].values
y = dataset['Consommation (L)'].values

# Créer et entraîner le modèle
model = LinearRegression()
model.fit(X, y)

# Prédire les valeurs avec le modèle
y_pred = model.predict(X)

plt.scatter(X, y, color='blue', label='Données observées')
plt.plot(X, y_pred, color='red', label='Régression Linéaire')

plt.xlabel("Température (°C)")
plt.ylabel("Consommation de boissons fraîches (L)")
plt.title("Température _ Consommation")
plt.legend()
plt.show()

# Afficher les coefficients du modèle
print("Pente (coefficient) :", model.coef_[0])
print("Ordonnée à l'origine :", model.intercept_)

# Calcul des erreurs
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Afficher les résultats d'évaluation
print("Erreur Absolue Moyenne (MAE) :", mae)
print("Erreur Quadratique Moyenne (MSE) :", mse)
print("Score R² :", r2)
```
