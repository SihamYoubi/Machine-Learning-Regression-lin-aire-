# Programme de nuage de points avec sklearn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Fixer la graine aléatoire pour la reproductibilité
np.random.seed(44)

# Définir les coefficients de la droite réelle
a = 2
b = 3

# Génération des données
nombre_points = 100
X = np.random.rand(nombre_points) * 10 
bruit = np.random.randn(nombre_points) * 2  # bruit gaussien avec écart type 2
y = a * X + b + bruit 

# Affichage du nuage de points
plt.scatter(X, y, label='Données')

# Remodeler X pour l'adapter au modèle
X = np.reshape(X, (len(X), 1))

# Créer le modèle de régression linéaire
model = LinearRegression()

# Entraîner le modèle
model.fit(X, y)

# Prédire les valeurs avec le modèle entraîné
y_pred = model.predict(X)

plt.plot(X, y_pred, color='red', label='Régression Linéaire')

plt.xlabel("X")
plt.ylabel("y")
plt.title("Régression Linéaire sur un Nuage de Points")
plt.legend()

plt.show()

# Afficher les coefficients appris par le modèle
print("Pente (coefficient a) :", model.coef_[0])
print("Ordonnée à l'origine (biais) :", model.intercept_)

# Calcul des erreurs
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Affichage des erreurs
print("Erreur Absolue Moyenne (MAE) :", mae)
print("Erreur Quadratique Moyenne (MSE) :", mse)
print("Score R² :", r2)
```
