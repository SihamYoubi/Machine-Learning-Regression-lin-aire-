# Programme de nuage de points sans sklearn

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Génération du dataset (nuage de points)
np.random.seed(44)
X = np.random.rand(100) * 10  
bruit = np.random.randn(100) * 2  
y = 2 * X + 3 + bruit  

X = X.reshape(-1, 1)

# 2. Split manuel des données (80% train, 20% test)
def split_data(X, y, test_size=0.2):
    n = len(X)
    indices = np.arange(n)
    np.random.shuffle(indices)
    test_count = int(n * test_size)
    test_idx = indices[:test_count]
    train_idx = indices[test_count:]
    return X[train_idx], X[test_idx], y[train_idx], y[test_idx]

X_train, X_test, y_train, y_test = split_data(X, y)

# 3. Implémentation du modèle linéaire avec gradient descent
def fit(X, y, learning_rate=0.01, epochs=1000):
    n = len(y)
    w = 0
    bias = 0
    mse_list = []

    for _ in range(epochs):
        y_pred = bias + w * X.flatten()
        error = y - y_pred
        mse = np.mean(error ** 2)
        mse_list.append(mse)

        # Calcul des gradients
        grad_bias = -(2/n) * np.sum(error)
        grad_w = -(2/n) * np.sum(X.flatten() * error)

        # Mise à jour des paramètres
        bias -= learning_rate * grad_bias
        w -= learning_rate * grad_w

    return bias, w, mse_list

bias, w, mse_list = fit(X_train, y_train)

# 4. Prédictions et évaluation
y_pred_test = bias + w * X_test.flatten()
mse_test = np.mean((y_test - y_pred_test) ** 2)
print(f"Paramètres entraînés: w = {w:.3f}, bias = {bias:.3f}")
print(f"MSE sur test set: {mse_test:.3f}")

# 5. Visualisation
plt.figure(figsize=(8, 6))
plt.scatter(X_train, y_train, color='blue', label='Train')
plt.scatter(X_test, y_test, color='red', label='Test')
plt.plot(X_test, y_pred_test, color='green', linewidth=2, label='Modèle')
plt.title('Régression Linéaire - Nuage de points (sans sklearn)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
